5. Incomplete Leads Reduction Over Time: Calculate the trend (using a linear regression model) of the number of incomplete leads over time for each associate. Are there any significant improvements or deteriorations?

```python
# Deleting previously created temporary DataFrame - associates
del associates
```

```python
# Temporary Dataframe containing 'No. of Incomplete Leads' of each associate under their name
associates = pd.DataFrame(columns= ["Raj", 'Ali', 'Arya', 'Date'])
associates['Raj'] = raj['No. of Incomplete Leads'].copy()
associates['Ali'] = ali['No. of Incomplete Leads'].copy()
associates['Arya'] = arya['No. of Incomplete Leads'].copy()
associates['Date'] = raj['Date'].copy()      #Since we know the three associates worked on same dates
associates
```

```python
# Checking for sum of NULL values
associates.isnull().sum()
```

```python
# Dropping rows (axis = 0) having null values. 'inplace = True' - to reflect changes in 'associates' dataframe
associates.dropna(axis=0, inplace=True)
```

```python
# Checking information about our temporary dataframe. 
# We can see there are 40 entries and all non-null along with their data types.
associates.info()
```

```python
# Changing format of Date (Required for Linear Regression modelling)
associates['Date'] = pd.to_datetime(associates['Date'], format='%d/%m/%Y')
```

```python
# 'x' - Number of Day(s) that has been worked by associates and reshaping the array into 2D, required for Linear Regression modelling.
x = np.array((associates['Date'] - associates['Date'].min()).dt.days).reshape(-1, 1)
y_raj = associates['Raj']
y_arya = associates['Arya']
y_ali = associates['Ali']
```

```python
# Individual models for associates
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()

model1.fit(x, y_raj)
model2.fit(x, y_arya)
model3.fit(x, y_ali)
```

```python
pred_raj = model1.predict(x)
pred_arya = model2.predict(x)
pred_ali = model3.predict(x)
```

```python
slope_raj = model1.coef_[0]
slope_arya = model2.coef_[0]
slope_ali = model3.coef_[0]
```

```python
plt.figure(figsize=(10, 4))
plt.plot(associates['Date'], pred_raj, color = associates_color[0])
plt.scatter(associates['Date'], y_raj, color = 'lightblue')
plt.title('Incomplete Leads Reduction Over Time \n(Raj)')
plt.text(associates['Date'].min(), max(y_raj), f'Slope: {slope_raj:.2f}');
```

```python
plt.figure(figsize=(10, 4))
plt.plot(associates['Date'], pred_ali, color = associates_color[1])
plt.scatter(associates['Date'], y_ali)
plt.title('Incomplete Leads Reduction Over Time \n(Ali)')
plt.text(associates['Date'].min(), max(y_ali)-0.25, f'Slope: {slope_ali:.2f}');
```

```python
plt.figure(figsize=(10, 4))
plt.plot(associates['Date'], pred_arya, color = associates_color[2])
plt.scatter(associates['Date'], y_arya)
plt.title('Incomplete Leads Reduction Over Time \n(Arya)')
plt.text(associates['Date'].min(), max(y_arya), f'Slope: {slope_arya:.2f}');
```

We can see that there's an overall reduction in the Incomplete Leads over time for each associate.

10. Predictive Analysis: Using a simple linear regression model, predict the number of leads each associate is expected to generate based on their time spent on lead generation. How accurate is the model when compared to actual data?

```python
#Creating new dataframe for training the model. It will contain columns Leads and Time spent for all associates - Raj, Arya and Ali.
model_df = pd.DataFrame(columns=['leads_generated_by_Raj', 'time_spent_by_Raj', 'average_time_by_Raj',
                                 'leads_generated_by_Arya', 'time_spent_by_Arya', 'average_time_by_Arya',
                                 'leads_generated_by_Ali', 'time_spent_by_Ali', 'average_time_by_Ali'
                                ])

# Copying values from orginal datframe to temporary dataframe for modelling
model_df['leads_generated_by_Raj'] = raj['Leads'].copy()
model_df['leads_generated_by_Arya'] = arya['Leads'].copy()
model_df['leads_generated_by_Ali'] = ali['Leads'].copy()

model_df['time_spent_by_Raj'] = raj['Time spent on LG (mins)'].copy()
model_df['time_spent_by_Arya'] = arya['Time spent on LG (mins)'].copy()
model_df['time_spent_by_Ali'] = ali['Time spent on LG (mins)'].copy()

model_df['average_time_by_Raj'] = raj['Avg Time Per Lead (mins)'].copy()
model_df['average_time_by_Arya'] = arya['Avg Time Per Lead (mins)'].copy()
model_df['average_time_by_Ali'] = ali['Avg Time Per Lead (mins)'].copy()

model_df.info()
```

```python
# Dropped NULL rows (axis = 0) for modelling
model_df.dropna(axis = 0, inplace = True)
model_df.info() 
```

```python
# Checking DataFrame
model_df.head(3)
```

```python
# Fixing Independent values (x) and Dependent values (y)
# [ y = f(x) ]

x = model_df.drop(['leads_generated_by_Raj', 'leads_generated_by_Arya', 'leads_generated_by_Ali'], axis = 1) # Removing mentioned columns and feeding the rest
y = model_df[['leads_generated_by_Raj', 'leads_generated_by_Arya', 'leads_generated_by_Ali']]   # Feeding the columns mentioned 
```

```python
# Splitting dataset for Training and Testing the model
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.65, random_state=40)

# Training size (train_size = 0.65) -> 65% of the dataset, 
# 'random_state = 40' -> Fixed the random rows selection that'll be used for training everytime we re-run the cell. It resulted in more accuracy of model.
```
